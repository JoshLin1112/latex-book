```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import poisson


lamb = 1  

n = 20
N = 100

samples = np.random.poisson(lamb, (n, N))

sample_sums = np.sum(samples, axis=0)
plt.figure(figsize=(10, 5))
plt.subplot(1, 2, 1)
plt.hist(sample_sums, bins=range(0, max(sample_sums) + 1), density=True, alpha=0.7, label='Sum of 20 Poisson(1)')
x = range(0, max(sample_sums) + 1)
pmf = poisson.pmf(x, n * lamb)
plt.plot(x, pmf, 'ro-', label='Poisson(20)')
plt.title('Sum of 20 Poisson(1) Random Variables')
plt.xlabel('x')
plt.ylabel('Probability')
plt.legend()

plt.subplot(1, 2, 2)
x = np.arange(0, max(sample_sums) + 1)
pmf = poisson.cdf(x, n * lamb)
plt.plot(x, pmf, 'ro-', label='Poisson(20) CDF')

values, counts = np.unique(sample_sums, return_counts=True)
ecdf = np.cumsum(counts) / len(sample_sums)
plt.step(values, ecdf, label='Empirical CDF', where='post')

plt.title('CDF Comparison')
plt.xlabel('x')
plt.ylabel('Probability')
plt.legend()

plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import expon, chi2


lambda_param = 3
sample_count = 100
sample_size = 30

samples = np.random.exponential(scale=1/lambda_param, size=(sample_count, sample_size))

sample_sums = 2 * lambda_param * np.sum(samples, axis=1)
plt.figure(figsize=(10, 5))
plt.subplot(1, 2, 1)
plt.hist(sample_sums, bins=20, density=True, alpha=0.7, label='Random Sample')
x = np.linspace(0, max(sample_sums), 1000)
pdf = chi2.pdf(x, df=2*sample_size)
plt.plot(x, pdf, 'r', label='Chi-squared(60) PDF')
plt.title('Chi-squared(60) Distribution vs. Random Sample')
plt.xlabel('x')
plt.ylabel('Probability')
plt.legend()

plt.subplot(1, 2, 2)
x = np.arange(0, max(sample_sums) + 1)
cdf = chi2.cdf(x, 2*sample_size)
plt.plot(x, cdf, 'r', label='Empirical CDF')
values, counts = np.unique(sample_sums, return_counts=True)
ecdf = np.cumsum(counts) / len(sample_sums)
plt.step(values, ecdf, label='CDF', where='post')
plt.xlabel('x')
plt.ylabel('Probability')
plt.legend()
plt.title(f'CDF Comparison (N={sample_size})')
plt.savefig(f'D:/latex/images/chi2_30.eps', format="eps", dpi=300)
plt.show()
print(sample_sums.mean())
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import expon, norm

lambda_param = 2
result_mu=[] 
result_var=[]

sample_sizes = [20, 100, 1000, 100000]
sample_count =100
for sample_size in sample_sizes:

    samples = np.random.exponential(scale=1/lambda_param, size=(sample_count, sample_size))

    sample_means = np.mean(samples,axis=0)

    plt.hist(sample_means, bins=20, density=True, alpha=0.7, label=f'Sample Size = {sample_size}')


    mu = 1/ lambda_param
    sigma = np.round(np.sqrt((1 / (4 * sample_count))),4)

    x = np.linspace(mu - 3 * sigma, mu + 3 * sigma, 100)
    pdf = norm.pdf(x, loc=mu, scale=sigma)
    plt.plot(x, pdf, 'r', label=f'N({mu}, {np.round(sigma**2,4)}) PDF')

    plt.title(f'Sample Means Distribution (Sample Size = {sample_size})')
    plt.xlabel('Sample Mean')
    plt.legend()
    result_mu.append(np.mean(sample_means))
    result_var.append(np.var(sample_means))
    plt.savefig(f'D:/latex/images/clt_{sample_size}.eps', format="eps", dpi=300)
    plt.show()
print(result_mu,result_var) 
```
