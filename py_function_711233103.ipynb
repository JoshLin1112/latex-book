```python
import numpy as np 
import matplotlib.pyplot as plt

x = np.linspace(-3*np.pi, 3*np.pi, 100)
y =  np.sin(x) / x 
y1 = np.sin(x**2) / x
y2 = (np.sin(2*x))**2 / x**2
y3 = (1 - np.cos(x)) / x**2

fig, ax = plt.subplots(2, 2, figsize=(8,5))

ax[0, 0].plot(x, y, color='b')
ax[0, 0].set_xticks(np.array([-3, -2, -1, 0, 1, 2, 3])*np.pi)
ax[0, 0].set_xticklabels(['-3$\pi$', '-2$\pi$', '$\pi$', '0', '$\pi$','2$\pi$','3$\pi$'], \
        fontsize=10, color='b')
ax[0, 0].set_ylim([-1, 1.5])
ax[0, 0].grid(True)
ax[0, 0].set_title('$y=\\frac{\\sin{x}}{x}$')

ax[0, 1].plot(x, y1, color='r')
ax[0, 1].set_xticks(np.array([-3, -2, -1, 0, 1, 2, 3])*np.pi)
ax[0, 1].set_xticklabels(['-3$\pi$', '-2$\pi$', '$\pi$', '0', '$\pi$','2$\pi$','3$\pi$'], \
        fontsize=10, color='r')
ax[0, 1].set_ylim([-1, 1])
ax[0, 1].grid(True)
ax[0, 1].set_title('$y=\\frac{\\sin{x^2}}{x}$')

ax[1, 0].plot(x, y2, color='g')
ax[1, 0].set_xticks(np.array([-3, -2, -1, 0, 1, 2, 3])*np.pi)
ax[1, 0].set_xticklabels(['-3$\pi$', '-2$\pi$', '$\pi$', '0', '$\pi$','2$\pi$','3$\pi$'], \
        fontsize=10, color='g')
ax[1, 0].set_ylim([-1, 5])
ax[1, 0].grid(True)
ax[1, 0].set_title('$y=\\frac{(\\sin{2x})^2}{x^2}$')

ax[1, 1].plot(x, y3, color='black')
ax[1, 1].set_xticks(np.array([-3, -2, -1, 0, 1, 2, 3])*np.pi)
ax[1, 1].set_xticklabels(['-3$\pi$', '-2$\pi$', '$\pi$', '0', '$\pi$','2$\pi$','3$\pi$'], \
        fontsize=10, color='black')
ax[1, 1].set_ylim([-0.5, 1])
ax[1, 1].grid(True)
ax[1, 1].set_title('$y=\\frac{1-\\cos{x}}{x^2}$')

plt.suptitle('Limit of Trigonometric Functions')
plt.subplots_adjust(hspace=0.4) 
plt.savefig("p1.eps", format="eps", dpi=300)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
a=1
x = np.arange(-8, 8, 0.1)
y = np.exp(a*x) / (1 + np.exp(a*x))

plt.figure(figsize=(8, 5))
plt.plot(x, y, color='b')
plt.axhline(y=1, color='r', linestyle='--')
plt.axhline(y=0, color='r', linestyle='--')

plt.title('$f(x) = \\frac{e^x}{1 + e^x}$')
plt.xlabel('x')
plt.ylabel('y')
plt.grid(True)
plt.text(-4, 0.4, '$f(x) = \\frac{e^x}{1 + e^x}$', fontsize=18, color='blue')
plt.text(5, 0.9, '$Asympote:y=1$', fontsize=15, color='r')
plt.text(-8, 0.1, '$Asympote:y=0$', fontsize=15, color='r')
plt.savefig("p2.eps", format="eps", dpi=300)
plt.show()
```

```python
#loop
alpha=1
n=5
x = np.arange(-6, 6, 0.1)
fig, ax = plt.subplots()
colors = ['b', 'c' ,'r', 'g', 'm']
for alpha in range(1,n+1):
    y = np.exp(alpha * x) / ( 1 + np.exp(alpha * x))
    plt.plot(x, y, label = f'$\\alpha={alpha}$', color = colors[alpha-1])

plt.legend()
plt.title('$f(x) = \\frac{e^{\\alpha x}}{1 + e^{\\alpha x}}$')
plt.xlabel('x')
plt.ylabel('y')
plt.grid(True)  
plt.savefig("p2-1.eps", format="eps", dpi=300)
plt.show()
```

```python
#broadcasting
n=10
x = np.arange(-6, 6, 0.1)
alpha = np.arange(1, n+1 , 1)
y = np.exp( x.reshape(-1 , 1) * alpha) / ( 1 + np.exp(x.reshape(-1 , 1) * alpha))
plt.plot(x, y, color = 'b', alpha = 0.5)
plt.xlabel('x')
plt.ylabel('y')
plt.grid(True)  
plt.title('$f(x) = \\frac{e^{\\alpha x}}{1 + e^{\\alpha x}}$')
plt.savefig("p2-2.eps", format="eps", dpi=300)
plt.show()
```

```python
from scipy.interpolate import make_interp_spline
x = np.arange(-40, 10, 0.1)
y = np.exp(-x/10) * np.sin(x)
a = np.array([-40, -30, -20, 0, 10]) 
b = np.array([-60, -30, -17, -5,-2])
c = np.array([60, 30, 17, 5,2])
curve1 = make_interp_spline(a, b)
curve2 = make_interp_spline(a, c)
ys=curve1(x)
zs=curve2(x)

plt.figure(figsize=(8, 5))
plt.plot(x, y,  color='b')
plt.plot(x, ys, x, zs, linestyle='--',color='gray')
plt.xlabel('x')
plt.ylabel('y')
plt.grid(True)
plt.text(0, -20 ,'$y=e^{-\\frac{x}{10}}\sin(x)$', color='b',fontsize=14)
plt.title('$y=e^{-\\frac{x}{10}}\sin(x)$')
plt.savefig("p3.eps", format="eps", dpi=300)
plt.show()
```

```python
avg = 3
sigma_square = 2
sigma = np.sqrt(sigma_square)
x = np.linspace(-2, 8 ,100)
y = (1/np.pi)*(np.exp((-(x-avg)**2)/(sigma_square)))

fig, ax = plt.subplots()
plt.plot(x ,y , color='r', linewidth = 3)
plt.axvline(x = avg, linestyle='dotted')

x_values = [avg-3*sigma, avg-2*sigma, avg-1*sigma, avg+sigma, avg+2*sigma, avg+3*sigma]
colors = ['gray', 'gray', 'gray','gray', 'gray', 'gray' ]
ax.set_xticks([avg-3*sigma,avg-2*sigma, avg-1*sigma,avg, avg+sigma, avg+2*sigma, avg+3*sigma])
ax.set_ylim([0,0.5])
for x, color in zip(x_values, colors):
    plt.axvline(x=x, color=color, linestyle='-.',linewidth=0.5)
arrow_properties = dict(arrowstyle='<->', color='g', lw=1*sigma)
ax.annotate('', xy=(1.586, 0.35), xytext=(4.414, 0.35), arrowprops=arrow_properties)
arrow_properties = dict(arrowstyle='<->', color='g', lw=1*sigma)
ax.annotate('', xy=(0.172, 0.4), xytext=(5.828, 0.4), arrowprops=arrow_properties)
arrow_properties = dict(arrowstyle='<->', color='g', lw=1*sigma)
ax.annotate('', xy=(-1.243, 0.45), xytext=(7.243, 0.45), arrowprops=arrow_properties)

plt.text(2.75,0.365,'64%')
plt.text(2.75,0.415,'95%')
plt.text(2.75,0.465,'99%')
plt.text(3.3, 0.13 , '$\\mu+\\sigma$')
plt.text(1.85, 0.13 , '$\\mu-\\sigma$')
plt.text(4.6, 0.13 , '$\\mu+2\\sigma$')
plt.text(0.35, 0.13 , '$\\mu-2\\sigma$')
plt.text(6, 0.13 , '$\\mu+3\\sigma$')
plt.text(-1, 0.13 , '$\\mu-3\\sigma$')
plt.text(4.4,0.25,'$N(3,4)$',fontsize=15, color='r')
plt.title('$Normal \ Distribution \ with \ \\mu=3 ,\ \\sigma^2=2$')
plt.xlabel('x')
plt.ylabel('y')
plt.savefig("p6.eps", format="eps", dpi=300)
plt.show()
```

```python
x = np.arange(-2 ,4 ,0.01)
f = lambda x : 3*x**3 - x**4
x_roots = [0,3]  
y_roots = [f(x) for x in x_roots]
x_max = 2.25
y_max = f(2.25)

fig,ax = plt.subplots(1, 2, figsize=(12, 4)) 
ax[0].plot(x, f(x), color='b')
ax[0].scatter(x_roots,y_roots, color='red', marker='o', label='roots')
ax[0].scatter(x_max, y_max, color='black', marker='o', label='maximum')
ax[0].set_ylim(-40, 25) 
ax[0].legend()
ax[0].set_xlabel('x')
ax[0].set_ylabel('y')
ax[0].text(0, -3 ,'(0,0)')
ax[0].text(3.1, -2 ,'(3,0)')
ax[0].text( 2.3 , 10.5 ,'(2.25,8.5)')
ax[0].set_title('$y=3x^3-x^4$ (Correct)')
ax[0].grid(True)

#-1~1
ax[1].plot(x, f(x), color='b')
ax[1].set_xlim(-1, 1) 
ax[1].set_ylim(-5, 5)  
ax[1].scatter(x_roots,y_roots, color='red', marker='o', label='roots')
ax[1].legend()
ax[1].set_xlabel('x')
ax[1].set_ylabel('y')
ax[1].set_title('$y=3x^3-x^4$ (Wrong)')
ax[1].text(0, -0.5 ,'(0,0)')
ax[1].grid(True)
plt.savefig("p7.eps", format="eps", dpi=300)
plt.show()
```

```python
ns = [1, 10, 100, 1000,10**5,10**6] 
result = [] 

for n in ns:
    Sn = 0
    Sn_values = []
    for k in range(1, n + 1):
        Sn = Sn+ 1 / k
        Sn_values.append(Sn)
    result.append(Sn_values)

plt.figure(figsize=(8, 5))
for i in range(len(ns)):
    plt.plot(range(1, ns[i] + 1), result[i])

for i, n in enumerate(ns):
    plt.scatter(n, result[i][-1], c='red')
    plt.annotate(f'n={n}', (n, result[i][-1]), textcoords="offset points", xytext=(0,10), ha='center')

plt.xlabel('n')
plt.ylabel('$S_n$')
plt.title('Show : $S_n$ is divergent')
plt.grid(True)
plt.savefig("pj1.eps", format="eps", dpi=300)
plt.show()
```

```python
a, b, n = 1, 10, 10
x1 = np.linspace(a, b, n)
x2 = np.linspace(0.5, 11, 100)
y1 = 1 / x1
y2 = 1 / x2

plt.bar(x1, y1, 1, align='edge', color='red', alpha=0.5)
plt.plot(x2, y2)
plt.fill_between(x2, y2, 0 , where =y2 > 0, color = 'white')
plt.xlabel('x')
plt.xticks(range(1,11,1))
plt.title('$\gamma_n=S_n-\ln (n+1)$')
plt.text(2, 1.5,'$f(x)=\\frac{1}{x}$')
plt.text(5,1, '$\gamma_n$ = shaded area')
plt.savefig("pj2.eps", format="eps", dpi=300)
plt.show()
```

```python
n= np.arange(1, 101)
S_n = np.cumsum(1 / n) 
gamma_n = S_n - np.log(n + 1)  
y_n = 1/2*(1-1/(n+1))

plt.plot(n, gamma_n, label='$\gamma_n = S_n - \ln(n+1)$')
plt.plot(n, y_n, label='$\\frac{1}{2}(1-\\frac{1}{n+1})$')
plt.axhline(1, xmin=0, xmax=100, label='1',color='green')

plt.xlabel('n')
plt.title('Comparison')
plt.grid(True)
plt.savefig("pj3.eps", format="eps", dpi=300)
plt.legend()
plt.show()
```
